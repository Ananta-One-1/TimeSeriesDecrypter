```python
from matplotlib import pyplot as plt
import json
with open("data.json") as f:
    data = json.load(f)
    plt.figure(figsize=(12, 12))
    plt.title("Original vs Reconstructed Signal")
    plt.plot(data["originalTimeSeries"], label="Original")
    plt.plot(data["combinedTimeSeries"], label="Reconstructed")
    plt.legend()
    plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.title("Progression from start point to end point.")
xcoordinate = [x := coordinate["coodrinates"]["1"] for coordinate in data["progression"]]
ycoordinate = [x := coordinate["coodrinates"]["2"] for coordinate in data["progression"]]
plt.plot(xcoordinate, ycoordinate)
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.scatter(xcoordinate, ycoordinate, c="orange", alpha=0.5)
plt.scatter(x=data["optimal"]["coodrinates"]["1"], y=data["optimal"]["coodrinates"]["2"], c="blue")
```

```python
plt.figure(figsize=(12, 12))
plt.plot(data["error"])
plt.title("Loss over iterations")
plt.ylabel("Loss")
plt.xlabel("Iterations")
```
